```python
# %% [markdown]
# # AI-Driven Integration of Solution Biophysics, Structural Modeling, and Validation
# ## Comprehensive Biomolecular Characterization Framework
#
# **Authors:** Research Team
# **Date:** August 2025
# **Version:** 1.0
#
# ### Abstract
# This notebook presents a comprehensive TensorFlow-based framework that unifies solution biophysics data,
# AI-driven structure prediction (AlphaFold), cryo-EM refinement, and PDB deposition. The framework demonstrates
# advanced machine learning models for analyzing synthetic biophysical data and implements a multimodal neural
# network for predicting hydrodynamic properties from integrated datasets.

# %% [markdown]
# ## 1. Environment Setup and Dependencies

# %%
# Core scientific computing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from scipy.stats import bootstrap
from scipy.optimize import minimize
import warnings
warnings.filterwarnings('ignore')

# Machine Learning and Deep Learning
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from sklearn.model_selection import train_test_split, StratifiedKFold, cross_val_score
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import (accuracy_score, precision_score, recall_score, f1_score,
                           mean_absolute_error, classification_report, confusion_matrix)
from sklearn.metrics.pairwise import cosine_similarity

# Bioinformatics and molecular analysis
import io
import requests
from datetime import datetime
import time

# Set random seeds for reproducibility
np.random.seed(42)
tf.random.set_seed(42)

# GPU configuration
gpus = tf.config.experimental.list_physical_devices('GPU')
if gpus:
    try:
        for gpu in gpus:
            tf.config.experimental.set_memory_growth(gpu, True)
        print(f"âœ… GPU acceleration enabled: {len(gpus)} GPU(s) available")
    except RuntimeError as e:
        print(f"âš ï¸ GPU configuration error: {e}")
else:
    print("â„¹ï¸ Running on CPU")

print("ðŸ§¬ Biomolecular Analysis Framework Initialized")
print(f"TensorFlow version: {tf.__version__}")
print(f"Pandas version: {pd.__version__}")
print(f"NumPy version: {np.__version__}")

# %% [markdown]
# ## 2. Core Biomolecular Analysis Framework

# %%
class BiomolecularAnalysisFramework:
    """
    AI-Driven Integration Framework for Solution Biophysics, Structural Modeling, and Validation
    """

    def __init__(self):
        self.scaler = StandardScaler()
        self.label_encoder = LabelEncoder()
        self.models = {}
        self.results = {}
        self.synthetic_data = {}
        print("ðŸ”¬ BiomolecularAnalysisFramework initialized")

    def generate_synthetic_biophysics_data(self, n_samples=5000):
        """
        Generate synthetic biophysics data simulating AUC, light scattering, and mass photometry
        """
        print(f"ðŸ“Š Generating synthetic biophysics datasets for {n_samples} samples...")

        # AUC time-series data (sedimentation velocity profiles)
        time_points = np.linspace(0, 100, 150)  # 150 time points
        auc_data = []

        for i in range(n_samples):
            # Simulate different protein oligomerization states
            oligomer_state = np.random.choice([1, 2, 4, 8])  # monomer, dimer, tetramer, octamer
            base_coeff = np.random.normal(2.5 + np.log(oligomer_state), 0.3)

            # Generate sedimentation profile with noise
            profile = np.exp(-time_points/base_coeff) * np.random.normal(1, 0.1, len(time_points))
            profile += np.random.normal(0, 0.05, len(time_points))  # measurement noise
            auc_data.append(profile)

        auc_data = np.array(auc_data)

        # Light scattering / Mass photometry data
        molecular_weights = np.random.lognormal(np.log(50000), 0.5, n_samples)  # 50kDa average
        hydrodynamic_radius = 2.1 * (molecular_weights ** 0.35) + np.random.normal(0, 0.5, n_samples)

        # AlphaFold confidence scores and structural features
        alphafold_confidence = np.random.beta(8, 2, n_samples)  # High confidence bias
        disorder_content = np.random.beta(2, 8, n_samples)  # Low disorder bias
        secondary_structure = np.random.dirichlet([3, 2, 1], n_samples)  # [helix, sheet, coil]

        # Validation metrics (simulating PDB validation)
        ramachandran_outliers = np.random.poisson(2, n_samples)
        clash_score = np.random.exponential(5, n_samples)

        # Temperature and time dependencies (non-linear effects)
        temperature = np.random.uniform(4, 37, n_samples)  # Celsius
        measurement_time = np.random.uniform(0, 24, n_samples)  # hours

        # Non-linear temperature effects on stability
        temp_stability = np.exp(-(temperature - 25)**2 / 200)
        time_degradation = np.exp(-measurement_time / 12)  # 12-hour half-life

        self.synthetic_data = {
            'auc_profiles': auc_data,
            'molecular_weight': molecular_weights,
            'hydrodynamic_radius': hydrodynamic_radius,
            'alphafold_confidence': alphafold_confidence,
            'disorder_content': disorder_content,
            'secondary_structure': secondary_structure,
            'ramachandran_outliers': ramachandran_outliers,
            'clash_score': clash_score,
            'temperature': temperature,
            'measurement_time': measurement_time,
            'temp_stability': temp_stability,
            'time_degradation': time_degradation,
            'oligomer_state': [np.random.choice([1, 2, 4, 8]) for _ in range(n_samples)]
        }

        print(f"âœ… Generated synthetic data: {len(self.synthetic_data)} feature types")
        return self.synthetic_data

    def perform_statistical_analysis(self):
        """
        Perform ANOVA, K-means clustering, Bootstrapping, and PCA analysis
        """
        print("\nðŸ“ˆ Performing comprehensive statistical analyses...")

        data = self.synthetic_data

        # 1. ANOVA Analysis
        print("1ï¸âƒ£ ANOVA Analysis:")
        oligomer_groups = []
        for state in [1, 2, 4, 8]:
            mask = np.array(data['oligomer_state']) == state
            oligomer_groups.append(data['hydrodynamic_radius'][mask])

        f_stat, p_value = stats.f_oneway(*oligomer_groups)
        print(f"   F-statistic: {f_stat:.4f}, p-value: {p_value:.4e}")

        # 2. K-means Clustering
        print("2ï¸âƒ£ K-means Clustering:")
        features_for_clustering = np.column_stack([
            data['molecular_weight'],
            data['hydrodynamic_radius'],
            data['alphafold_confidence'],
            data['disorder_content']
        ])

        scaler = StandardScaler()
        features_scaled = scaler.fit_transform(features_for_clustering)

        kmeans = KMeans(n_clusters=4, random_state=42, n_init=10)
        cluster_labels = kmeans.fit_predict(features_scaled)

        silhouette_score = self._calculate_silhouette_score(features_scaled, cluster_labels)
        print(f"   Silhouette Score: {silhouette_score:.4f}")

        # 3. Bootstrap Analysis
        print("3ï¸âƒ£ Bootstrap Analysis:")
        def mean_confidence(x):
            return np.mean(x)

        rng = np.random.default_rng(42)
        res = bootstrap((data['hydrodynamic_radius'],), mean_confidence,
                       n_resamples=1000, random_state=rng)

        bootstrap_ci = res.confidence_interval
        print(f"   Bootstrap 95% CI for mean hydrodynamic radius: [{bootstrap_ci.low:.2f}, {bootstrap_ci.high:.2f}]")

        # 4. PCA Analysis
        print("4ï¸âƒ£ PCA Analysis:")
        pca = PCA(n_components=4)
        pca_result = pca.fit_transform(features_scaled)

        explained_variance = pca.explained_variance_ratio_
        print(f"   Explained variance ratio: {explained_variance}")
        print(f"   Cumulative explained variance: {np.cumsum(explained_variance)}")

        self.results['statistical_analysis'] = {
            'anova': {'f_stat': f_stat, 'p_value': p_value},
            'kmeans': {'labels': cluster_labels, 'silhouette': silhouette_score, 'centers': kmeans.cluster_centers_},
            'bootstrap_ci': bootstrap_ci,
            'pca': {'components': pca_result, 'explained_variance': explained_variance, 'loadings': pca.components_}
        }

        print("âœ… Statistical analysis completed")
        return self.results['statistical_analysis']

    def _calculate_silhouette_score(self, X, labels):
        """Calculate silhouette score manually"""
        n_samples = len(X)
        silhouette_scores = []

        for i in range(n_samples):
            same_cluster = labels == labels[i]
            other_clusters = labels != labels[i]

            if np.sum(same_cluster) == 1:
                silhouette_scores.append(0)
                continue

            # Mean distance to points in same cluster (excluding self)
            same_cluster[i] = False  # Exclude self
            if np.sum(same_cluster) == 0:
                a = 0
            else:
                a = np.mean([np.linalg.norm(X[i] - X[j]) for j in range(n_samples) if same_cluster[j]])

            # Mean distance to points in nearest other cluster
            b = float('inf')
            for cluster_id in np.unique(labels[other_clusters]):
                cluster_mask = labels == cluster_id
                cluster_distances = [np.linalg.norm(X[i] - X[j]) for j in range(n_samples) if cluster_mask[j]]
                if cluster_distances:
                    mean_dist = np.mean(cluster_distances)
                    b = min(b, mean_dist)

            if b == float('inf') or max(a, b) == 0:
                silhouette_scores.append(0)
            else:
                s = (b - a) / max(a, b)
                silhouette_scores.append(s)

        return np.mean(silhouette_scores)

# Initialize framework
framework = BiomolecularAnalysisFramework()

# %% [markdown]
# ## 3. Synthetic Data Generation and Visualization

# %%
# Generate synthetic biophysics data
synthetic_data = framework.generate_synthetic_biophysics_data(n_samples=5000)

# Create comprehensive visualization
fig, axes = plt.subplots(2, 4, figsize=(20, 10))
fig.suptitle('ðŸ§¬ Synthetic Biophysics Data Overview', fontsize=16, fontweight='bold')

# 1. AUC Profile Examples
for i in range(5):
    axes[0, 0].plot(synthetic_data['auc_profiles'][i], alpha=0.7, label=f'Sample {i+1}')
axes[0, 0].set_title('AUC Sedimentation Profiles')
axes[0, 0].set_xlabel('Time Points')
axes[0, 0].set_ylabel('Signal Intensity')
axes[0, 0].legend()

# 2. Molecular Weight Distribution
axes[0, 1].hist(synthetic_data['molecular_weight']/1000, bins=50, alpha=0.7, color='skyblue', edgecolor='black')
axes[0, 1].set_title('Molecular Weight Distribution')
axes[0, 1].set_xlabel('Molecular Weight (kDa)')
axes[0, 1].set_ylabel('Frequency')

# 3. Temperature vs Hydrodynamic Radius
scatter = axes[0, 2].scatter(synthetic_data['temperature'], synthetic_data['hydrodynamic_radius'],
                           c=synthetic_data['molecular_weight'], cmap='viridis', alpha=0.6)
axes[0, 2].set_title('Temperature vs Hydrodynamic Radius')
axes[0, 2].set_xlabel('Temperature (Â°C)')
axes[0, 2].set_ylabel('Hydrodynamic Radius (nm)')
plt.colorbar(scatter, ax=axes[0, 2], label='Mol. Weight (Da)')

# 4. AlphaFold Confidence vs Disorder
scatter2 = axes[0, 3].scatter(synthetic_data['alphafold_confidence'], synthetic_data['disorder_content'],
                            c=synthetic_data['ramachandran_outliers'], cmap='Reds', alpha=0.6)
axes[0, 3].set_title('AlphaFold Confidence vs Disorder')
axes[0, 3].set_xlabel('AlphaFold Confidence')
axes[0, 3].set_ylabel('Disorder Content')
plt.colorbar(scatter2, ax=axes[0, 3], label='Ramachandran Outliers')

# 5. Secondary Structure Composition
ss_mean = np.mean(synthetic_data['secondary_structure'], axis=0)
ss_labels = ['Î±-Helix', 'Î²-Sheet', 'Coil']
axes[1, 0].pie(ss_mean, labels=ss_labels, autopct='%1.1f%%', startangle=90)
axes[1, 0].set_title('Average Secondary Structure')

# 6. Oligomerization State Distribution
oligomer_counts = pd.Series(synthetic_data['oligomer_state']).value_counts().sort_index()
axes[1, 1].bar(oligomer_counts.index, oligomer_counts.values, alpha=0.7, color='orange', edgecolor='black')
axes[1, 1].set_title('Oligomerization State Distribution')
axes[1, 1].set_xlabel('Oligomer State')
axes[1, 1].set_ylabel('Count')

# 7. Time-Temperature Correlation
axes[1, 2].scatter(synthetic_data['measurement_time'], synthetic_data['temperature'],
                  c=synthetic_data['temp_stability'], cmap='coolwarm', alpha=0.6)
axes[1, 2].set_title('Measurement Time vs Temperature')
axes[1, 2].set_xlabel('Measurement Time (hours)')
axes[1, 2].set_ylabel('Temperature (Â°C)')

# 8. Validation Metrics
axes[1, 3].scatter(synthetic_data['clash_score'], synthetic_data['ramachandran_outliers'],
                  c=synthetic_data['alphafold_confidence'], cmap='RdYlGn', alpha=0.6)
axes[1, 3].set_title('Structure Validation Metrics')
axes[1, 3].set_xlabel('Clash Score')
axes[1, 3].set_ylabel('Ramachandran Outliers')

plt.tight_layout()
plt.show()

print("âœ… Data visualization completed")

# %% [markdown]
# ## 4. Statistical Analysis Suite

# %%
# Perform comprehensive statistical analysis
stats_results = framework.perform_statistical_analysis()

# Visualize statistical results
fig, axes = plt.subplots(2, 2, figsize=(15, 12))
fig.suptitle('ðŸ“Š Statistical Analysis Results', fontsize=16, fontweight='bold')

# 1. ANOVA Results - Box plot by oligomer state
oligomer_data = []
oligomer_labels = []
for state in [1, 2, 4, 8]:
    mask = np.array(synthetic_data['oligomer_state']) == state
    oligomer_data.append(synthetic_data['hydrodynamic_radius'][mask])
    oligomer_labels.append(f'State {state}')

axes[0, 0].boxplot(oligomer_data, labels=oligomer_labels)
axes[0, 0].set_title(f'ANOVA Results\nF={stats_results["anova"]["f_stat"]:.2f}, p={stats_results["anova"]["p_value"]:.2e}')
axes[0, 0].set_xlabel('Oligomer State')
axes[0, 0].set_ylabel('Hydrodynamic Radius (nm)')

# 2. K-means Clustering Results
pca_components = stats_results['pca']['components']
cluster_labels = stats_results['kmeans']['labels']
scatter = axes[0, 1].scatter(pca_components[:, 0], pca_components[:, 1],
                           c=cluster_labels, cmap='Set1', alpha=0.6)
axes[0, 1].set_title(f'K-means Clustering (PCA space)\nSilhouette Score: {stats_results["kmeans"]["silhouette"]:.3f}')
axes[0, 1].set_xlabel('PC1')
axes[0, 1].set_ylabel('PC2')

# 3. PCA Explained Variance
explained_var = stats_results['pca']['explained_variance']
cumulative_var = np.cumsum(explained_var)
x_pos = range(1, len(explained_var) + 1)
bars = axes[1, 0].bar(x_pos, explained_var, alpha=0.7, color='steelblue', label='Individual')
axes[1, 0].plot(x_pos, cumulative_var, 'ro-', label='Cumulative')
axes[1, 0].set_title('PCA Explained Variance')
axes[1, 0].set_xlabel('Principal Component')
axes[1, 0].set_ylabel('Explained Variance Ratio')
axes[1, 0].legend()

# 4. Bootstrap Confidence Interval
bootstrap_ci = stats_results['bootstrap_ci']
mean_val = np.mean(synthetic_data['hydrodynamic_radius'])
axes[1, 1].hist(synthetic_data['hydrodynamic_radius'], bins=50, alpha=0.7, color='lightgreen', density=True)
axes[1, 1].axvline(mean_val, color='red', linestyle='--', linewidth=2, label=f'Mean: {mean_val:.2f}')
axes[1, 1].axvline(bootstrap_ci.low, color='orange', linestyle=':', linewidth=2, label=f'CI Low: {bootstrap_ci.low:.2f}')
axes[1, 1].axvline(bootstrap_ci.high, color='orange', linestyle=':', linewidth=2, label=f'CI High: {bootstrap_ci.high:.2f}')
axes[1, 1].set_title('Bootstrap Confidence Interval')
axes[1, 1].set_xlabel('Hydrodynamic Radius (nm)')
axes[1, 1].set_ylabel('Density')
axes[1, 1].legend()

plt.tight_layout()
plt.show()

print("âœ… Statistical analysis visualization completed")

# %% [markdown]
# ## 5. TensorFlow CNN Model Development

# %%
def build_tensorflow_cnn_model(framework):
    """
    Build advanced TensorFlow CNN model with non-linearity, time and temperature dependencies
    """
    print("ðŸ—ï¸ Building Advanced TensorFlow CNN Model...")

    # Input layers
    auc_input = keras.Input(shape=(150,), name='auc_profile')
    structural_input = keras.Input(shape=(7,), name='structural_features')  # AlphaFold + validation
    temporal_input = keras.Input(shape=(2,), name='temporal_features')  # temperature + time

    # CNN branch for AUC time-series data with attention
    auc_reshaped = layers.Reshape((150, 1))(auc_input)

    # Multi-scale convolutions
    conv_3 = layers.Conv1D(64, 3, activation='relu', padding='same')(auc_reshaped)
    conv_5 = layers.Conv1D(64, 5, activation='relu', padding='same')(auc_reshaped)
    conv_7 = layers.Conv1D(64, 7, activation='relu', padding='same')(auc_reshaped)

    # Concatenate multi-scale features
    multi_scale = layers.Concatenate()([conv_3, conv_5, conv_7])
    multi_scale = layers.BatchNormalization()(multi_scale)
    multi_scale = layers.Dropout(0.2)(multi_scale)

    # Attention mechanism
    attention = layers.Dense(192, activation='tanh')(multi_scale)
    attention = layers.Dense(192, activation='softmax')(attention)
    attended_features = layers.Multiply()([multi_scale, attention])

    # Global pooling and additional processing
    global_avg = layers.GlobalAveragePooling1D()(attended_features)
    global_max = layers.GlobalMaxPooling1D()(attended_features)
    global_features = layers.Concatenate()([global_avg, global_max])

    # LSTM for temporal dependencies
    lstm_input = layers.Reshape((150, 1))(auc_input)
    lstm = layers.LSTM(64, return_sequences=True, dropout=0.2, recurrent_dropout=0.2)(lstm_input)
    lstm = layers.LSTM(32, return_sequences=False, dropout=0.2)(lstm)

    # Residual blocks for structural features
    struct_dense1 = layers.Dense(128, activation='relu')(structural_input)
    struct_dense1 = layers.BatchNormalization()(struct_dense1)
    struct_dense1 = layers.Dropout(0.3)(struct_dense1)

    struct_dense2 = layers.Dense(128, activation='relu')(struct_dense1)
    struct_dense2 = layers.BatchNormalization()(struct_dense2)

    # Residual connection
    struct_residual = layers.Add()([struct_dense1, struct_dense2])
    struct_final = layers.Dense(64, activation='relu')(struct_residual)

    # Non-linear temporal processing
    temporal_dense1 = layers.Dense(32, activation='tanh')(temporal_input)  # tanh for temperature
    temporal_dense2 = layers.Dense(16, activation='sigmoid')(temporal_dense1)  # sigmoid for time decay
    temporal_dense3 = layers.Dense(8, activation='relu')(temporal_dense2)

    # Advanced fusion layer
    all_features = layers.Concatenate()([global_features, lstm, struct_final, temporal_dense3])

    # Deep fusion network with residual connections
    fusion1 = layers.Dense(256, activation='swish')(all_features)
    fusion1 = layers.BatchNormalization()(fusion1)
    fusion1 = layers.Dropout(0.3)(fusion1)

    fusion2 = layers.Dense(128, activation='swish')(fusion1)
    fusion2 = layers.BatchNormalization()(fusion2)
    fusion2 = layers.Dropout(0.2)(fusion2)

    # Residual connection in fusion
    fusion_residual = layers.Dense(128, activation='linear')(all_features)
    fusion_combined = layers.Add()([fusion2, fusion_residual])
    fusion_combined = layers.Activation('relu')(fusion_combined)

    fusion3 = layers.Dense(64, activation='relu')(fusion_combined)
    fusion3 = layers.BatchNormalization()(fusion3)

    # Multi-task outputs
    hydrodynamic_radius = layers.Dense(1, activation='linear', name='hydrodynamic_radius')(fusion3)
    molecular_weight = layers.Dense(1, activation='linear', name='molecular_weight')(fusion3)
    stability_score = layers.Dense(1, activation='sigmoid', name='stability_score')(fusion3)

    # Build model
    model = keras.Model(
        inputs=[auc_input, structural_input, temporal_input],
        outputs=[hydrodynamic_radius, molecular_weight, stability_score],
        name='AdvancedBiomolecularCNN'
    )

    # Custom physics-informed loss function
    def physics_informed_loss(y_true, y_pred):
        mse_loss = tf.keras.losses.MeanSquaredError()(y_true, y_pred)

        # Physical constraints
        negative_penalty = tf.reduce_mean(tf.nn.relu(-y_pred))  # Radius should be positive
        size_penalty = tf.reduce_mean(tf.nn.relu(y_pred - 50.0)) + tf.reduce_mean(tf.nn.relu(0.5 - y_pred))

        return mse_loss + 0.1 * negative_penalty + 0.1 * size_penalty

    # Compile with advanced optimizer
    model.compile(
        optimizer=keras.optimizers.AdamW(learning_rate=0.001, weight_decay=0.01),
        loss={
            'hydrodynamic_radius': physics_informed_loss,
            'molecular_weight': 'mse',
            'stability_score': 'binary_crossentropy'
        },
        loss_weights={
            'hydrodynamic_radius': 0.5,
            'molecular_weight': 0.3,
            'stability_score': 0.2
        },
        metrics={
            'hydrodynamic_radius': ['mae'],
            'molecular_weight': ['mae'],
            'stability_score': ['accuracy']
        }
    )

    framework.models['advanced_cnn'] = model
    print(f"âœ… Advanced CNN model built: {model.count_params():,} parameters")
    return model

# Build the CNN model
cnn_model = build_tensorflow_cnn_model(framework)

# Display model architecture
print("\nðŸ—ï¸ Model Architecture Summary:")
cnn_model.summary()

# %% [markdown]
# ## 6. Model Training with 40,000 Iterations

# %%
def train_advanced_cnn_model(framework, training_iterations=40000):
    """
    Train the advanced CNN model with specified iterations
    """
    print(f"ðŸš€ Training Advanced CNN model for {training_iterations} iterations...")

    data = framework.synthetic_data

    # Prepare training data
    X_auc = data['auc_profiles']
    X_structural = np.column_stack([
        data['alphafold_confidence'],
        data['disorder_content'],
        data['secondary_structure'],  # 3 columns
        data['ramachandran_outliers'],
        data['clash_score']
    ])
    X_temporal = np.column_stack([
        data['temperature'],
        data['measurement_time']
    ])

    # Multi-task targets
    y_radius = data['hydrodynamic_radius']
    y_weight = data['molecular_weight']
    y_stability = data['temp_stability']  # Use temperature stability as binary target
    y_stability_binary = (y_stability > np.median(y_stability)).astype(int)

    # Normalize features
    scaler_structural = StandardScaler()
    scaler_temporal = StandardScaler()
    scaler_weight = StandardScaler()

    X_structural_norm = scaler_structural.fit_transform(X_structural)
    X_temporal_norm = scaler_temporal.fit_transform(X_temporal)
    y_weight_norm = scaler_weight.fit_transform(y_weight.reshape(-1, 1)).flatten()

    # Split data
    indices = np.arange(len(y_radius))
    train_idx, test_idx = train_test_split(indices, test_size=0.2, random_state=42)
    val_idx, test_idx = train_test_split(test_idx, test_size=0.5, random_state=42)

    # Prepare training sets
    X_train = [
        X_auc[train_idx],
        X_structural_norm[train_idx],
        X_temporal_norm[train_idx]
    ]
    y_train = {
        'hydrodynamic_radius': y_radius[train_idx],
        'molecular_weight': y_weight_norm[train_idx],
        'stability_score': y_stability_binary[train_idx]
    }

    # Prepare validation sets
    X_val = [
        X_auc[val_idx],
        X_structural_norm[val_idx],
        X_temporal_norm[val_idx]
    ]
    y_val = {
        'hydrodynamic_radius': y_radius[val_idx],
        'molecular_weight': y_weight_norm[val_idx],
        'stability_score': y_stability_binary[val_idx]
    }

    # Prepare test sets
    X_test = [
        X_auc[test_idx],
        X_structural_norm[test_idx],
        X_temporal_norm[test_idx]
    ]
    y_test = {
        'hydrodynamic_radius': y_radius[test_idx],
        'molecular_weight': y_weight_norm[test_idx],
        'stability_score': y_stability_binary[test_idx]
    }

    # Calculate training parameters
    batch_size = 32
    total_samples = len(train_idx)
    steps_per_epoch = max(1, total_samples // batch_size)
    epochs = max(1, training_iterations // steps_per_epoch)

    print(f"ðŸ“Š Training configuration:")
    print(f"   Total iterations: {training_iterations}")
    print(f"   Batch size: {batch_size}")
    print(f"   Steps per epoch: {steps_per_epoch}")
    print(f"   Total epochs: {epochs}")
    print(f"   Training samples: {len(train_idx)}")
    print(f"   Validation samples: {len(val_idx)}")
    print(f"   Test samples: {len(test_idx)}")

    # Advanced callbacks
    callbacks = [
        keras.callbacks.EarlyStopping(
            patience=50,
            restore_best_weights=True,
            monitor='val_loss',
            verbose=1
        ),
        keras.callbacks.ReduceLROnPlateau(
            patience=20,
            factor=0.5,
            min_lr=1e-6,
            monitor='val_loss',
            verbose=1
        )
    ]

    # Train the model
    history = framework.models['advanced_cnn'].fit(
        X_train,
        y_train,
        epochs=epochs,
        batch_size=batch_size,
        validation_data=(X_val, y_val),
        callbacks=callbacks,
        verbose=1 # Changed to 1 to show progress
    )

    return history # Return the history object
```
